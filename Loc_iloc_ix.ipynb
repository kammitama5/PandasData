```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python

ufo.head()
```

```python
# loc is used for filtering rows and selecting columns by label
ufo.loc[0, :] # row zero, all columns
```

```python
ufo.loc[0:2, :]
```

```python
ufo.loc[0:2]
```

```python
ufo.loc[:, ['City', 'State']] # City, State
```

```python
ufo.loc[:, 'City':'State'] # city through State
```

```python
ufo.loc[0:2, 'City': 'State'] # rows 0 through 2 and City through State
```

```python
ufo.head(3).drop('Time', axis=1) 
# accomplishes same as
# ufo.loc[0:2, 'City': 'State'] # rows 0 through 2 and City through State
```

```python
ufo[ufo.City == 'Oakland']
# where City is Oakland
```

```python
ufo.loc[ufo.City == 'Oakland', :]
# , : means -> columns all
```

```python
ufo[ufo.City == 'Oakland'].State # chained indexing = not good
```

```python
ufo.loc[ufo.City == 'Oakland', 'State'] # this is better
```

```python
# iloc -> filtering by rows and columns by integer position
ufo.iloc[:, [0, 3]] # positions row and three in database
```

```python
ufo.iloc[:, 0:4] # range iloc is exclusive of second number, inclusive of first
# so shows rows 0 through 3 (excludes four)
```

```python
ufo.iloc[0:3, :] # rows 0, 1, 2 and all columns
```
